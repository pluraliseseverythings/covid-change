```python
import pandas
from IPython.display import display

CORONA_DATA_FILE = "COVID-19-geographic-disbtribution-worldwide-2020-03-24.xlsx"
corona_table = pandas.read_excel(CORONA_DATA_FILE)
corona_table = corona_table.groupby('GeoId').filter(lambda g: g.Cases.sum() > 1500)

# Cleaning up some data
ban = ['CANADA', 'Cases_on_an_international_conveyance_Japan']
corona_table = corona_table[~corona_table["Countries and territories"].isin(ban)]


```

```python
display(corona_table)
```

```python
import plotly.graph_objects as go
from plotly.offline import iplot
import plotly.express as px

fig = px.line(corona_table, x="DateRep", y="Cases", color="Countries and territories")
fig.show()
```

```python
corona_table = corona_table.sort_values(by=["Countries and territories", "DateRep"])

from collections import defaultdict
import numpy as np

country_partial_sum = defaultdict(int)
corona_table['Cumulative'] = corona_table.groupby(by=['Countries and territories'])['Cases'].cumsum()
corona_table['Growth'] = np.where( corona_table['Cumulative'] > 25, corona_table['Cases'].divide(corona_table['Cumulative']), float('nan'))
corona_table['Growth Rolling'] = corona_table['Growth'].rolling(10).mean()
fig = px.line(corona_table, x="DateRep", y="Cumulative", color="Countries and territories")
fig.show()
```

```python
fig = px.line(corona_table, x="DateRep", y="Growth Rolling", color="Countries and territories")
fig.show()
```

```python
fig = px.line(corona_table[corona_table["Countries and territories"].isin(["Australia", "Canada", "Portugal", "Brazil", "United_Kingdom", "United_States_of_America"])], x="DateRep", y="Growth Rolling", color="Countries and territories")
fig.update_layout(
    xaxis = dict(
        range=("2020-03-01", "2020-03-24"),
        constrain='domain'
    )
)
fig.show()
```

```python
lockdowns_before_20 = ["Denmark", "Czech_Republic", "Belgium", "Austria", "France", "Italy", "Ireland", "Malaysia", "Poland", "Spain"]
fig = px.line(corona_table[corona_table["Countries and territories"].isin(lockdowns_before_20)], x="DateRep", y="Growth Rolling", color="Countries and territories")
fig.update_layout(
    xaxis = dict(
        range=("2020-03-01", "2020-03-24"),
        constrain='domain'
    )
)
fig.show()
```
